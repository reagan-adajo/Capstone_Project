```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("clean_sales_data.csv")
df.head()
```

```python
df['sale_date'] = pd.to_datetime(df['sale_date'])
df['year'] = df['sale_date'].dt.year
```

```python
df.head()
```

```python
# Find negative numerical values in numerical columns
negative_values = (df.select_dtypes(include=['number']) < 0).sum()
negative_values
```

```python
# Filter rows where any numerical column has a negative value
df[df.select_dtypes(include=['number']).lt(0).any(axis=1)]
```

```python
# Drop rows where any numerical column has a negative value
df = df[~(df.select_dtypes(include=['number']) < 0).any(axis=1)]

# Reset index after dropping
df.reset_index(drop=True, inplace=True)
```

### Feature Engineering

#### 1. Time-Based Features

Sales data is affected by seasonality. Extracting time-based features helps capture these trends.

**How?**
- Day, Week, and Quarter: Help identify cyclical patterns.
- Weekend Indicator: Certain products see higher demands on weekends.

```python
df.loc[:, 'week'] = df['sale_date'].dt.isocalendar().week
df.loc[:, 'quarter'] = df['sale_date'].dt.quarter
df.loc[:, 'weekend'] = df['day_of_week'].isin([5, 6]).astype(bool)  # Explicitly cast to boolean
```

```python
df.head()
```

### 2. Lag Features (Previous Sales Data As Features)

**What are Lag Features**

Lag features are past values of a time-series variable used as input features for a machine learning model to predict future values. They help capture temporal dependencies in data, making them crucial for time-series forecasting, such as sales demand prediction.

**Why?**

Past sales often influence future demand. If a product sold well last week, it may sell well this week too.

**How?**

- **Lagging 1 day:** Captures immediate trends (yesterday's high sales may indicate today's demand).
- **Lagging 7 days:** Captures weekly seasonality (demand spikes every Friday).
- **Lagging 30 days:** Captures monthly trends (salary-based purchases at month-end). 

```python
df.loc[:, 'quantity_lag_1'] = df.groupby(['item_id', 'store_id'])['quantity'].shift(1).fillna(0)
df.loc[:, 'quantity_lag_7'] = df.groupby(['item_id', 'store_id'])['quantity'].shift(7).fillna(0)
df.loc[:, 'quantity_lag_30'] = df.groupby(['item_id', 'store_id'])['quantity'].shift(30).fillna(0)
```

```python
df.sample(10)
```

- `shift(n)` moves values downwards by `n` steps within each group.
- The first `n` rows in each group have no past data, so Pandas fills them with `NaN`.
- `fillna(0)` replaces the `NaN` values with `0`.

### 3. Rolling Statistics (Trend & Seasonality)


**Why?**

Instead of relying on a single day's past sales, rolling statistics smooths out trends and highlights volatility.

**How?**

- **Rolling Mean:** Shows the average demand over a period (eg, last 7 days).
- **Rolling Std:** Measures demand fluctuations (eg, is demand stable or highly variable?)
- **Expanding Mean:** Tracks long-term sales growth.

*Trends are more reliable when observed over time.* 

```python
df.loc[:, 'rolling_mean_7'] = df.groupby(['item_id', 'store_id'])['quantity'].transform(lambda x: x.shift(1).rolling(7).mean()).fillna(0)
df.loc[:, 'rolling_mean_14'] = df.groupby(['item_id', 'store_id'])['quantity'].transform(lambda x: x.shift(1).rolling(14).mean()).fillna(0)
df.loc[:, 'rolling_mean_30'] = df.groupby(['item_id', 'store_id'])['quantity'].transform(lambda x: x.shift(1).rolling(30).mean()).fillna(0)
df.loc[:, 'rolling_std_7'] = df.groupby(['item_id', 'store_id'])['quantity'].transform(lambda x: x.shift(1).rolling(7).std()).fillna(0)
df.loc[:, 'rolling_std_14'] = df.groupby(['item_id', 'store_id'])['quantity'].transform(lambda x: x.shift(1).rolling(14).std()).fillna(0)
df.loc[:, 'rolling_std_30'] = df.groupby(['item_id', 'store_id'])['quantity'].transform(lambda x: x.shift(1).rolling(30).std()).fillna(0)
df.loc[:, 'expanding_mean'] = df.groupby(['item_id', 'store_id'])['quantity'].transform(lambda x: x.expanding().mean())
```

- **Rolling Mean** helps capture local trends over a moving window(eg, weekly trends).
- **Rolling Std** captures sales volatility.
- **Expanding Mean** captures long-term trends.

### 4. Price & Discount Features

**Why?**

Price changes and discounts directly influence sales. Customers tend to buy more during promotions.

**How?**

- **Discount Rate:** Measures how much a product's price was reduced.
- **Discount Effect:** Shows the relationship between discounts and quantity sold.

*Customers Love Discounts*

```python
df.loc[:, 'discount_rate'] = df['discount'] / (df['effective_price'] + 1e-5) # Avoid division by 0
df.loc[:, 'discount_effect'] = df['quantity'] * df['discount_rate'] # How much discount influences sales
```

### 5. Store Features

```python
df.loc[:, 'store_avg_sales'] = df.groupby('store_id')['quantity'].transform('mean')
df.loc[:, 'city_avg_sales'] = df.groupby('city')['quantity'].transform('mean')
```
