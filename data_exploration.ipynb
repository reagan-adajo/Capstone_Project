```python
import pandas as pd

# Convert dates to datetime format
total_sales['sale_date'] = pd.to_datetime(total_sales['date'])
price_history['change_date'] = pd.to_datetime(price_history['date'])

# Merge to get the latest price before or on the sale date
price_sale_df = pd.merge_asof(
    total_sales.sort_values('sale_date'),
    price_history.sort_values('change_date'),
    left_on='sale_date',
    right_on='change_date',
    by=['item_id', 'store_id'],
    direction='backward'
)

# Create a new effective price column
price_sale_df['effective_price'] = price_sale_df['price'].combine_first(price_sale_df['price_base'])

# Automatically update sum_total using the new effective prices
price_sale_df['sum_total'] = price_sale_df['effective_price'] * price_sale_df['quantity']

# Check the updated dataframe
price_sale_df[['item_id', 'store_id', 'sale_date', 'quantity', 'effective_price', 'sum_total']].head()
```

`discounts_history.csv`

- **Purpose:** Contains historical promo data for each specific store.
- **Columns:**
    - `date`: Date
    - `item_id`: A unique identifier for each product
    - `sale_price_before_promo`: Price before promo period started
    - `sale_price_time_promo`: Price during the promo period
    - `promo_type_code`: Promo code type
    - `doc_id`: Promo document number
    - `number_disc_day`: Sequential day number of the current promo period
    - `store_id`: Store number

```python
discount_history = pd.read_csv('datasets/discounts_history.csv')
discount_history.head()
```

```python
discount_history.shape
```

```python
discount_history['item_id'].nunique()
```

```python
import pandas as pd

# Convert dates to datetime format
price_sale_df['sale_date'] = pd.to_datetime(price_sale_df['sale_date'])
discount_history['date'] = pd.to_datetime(discount_history['date'])

# Create a column for the promo end date
discount_history['promo_end_date'] = discount_history['date'] + pd.to_timedelta(discount_history['number_disc_day'], unit='D')

# Merge asof to get the last discount before each sale
merged_df = pd.merge_asof(
    price_sale_df.sort_values('sale_date'),
    discount_history.sort_values('date'),
    left_on='sale_date',
    right_on='date',
    by=['item_id', 'store_id'],
    direction='backward'  # Get the most recent discount before the sale
)


```

```python
# Check if the sale occurred within the discount period
merged_df['on_promo'] = (merged_df['sale_date'] >= merged_df['date']) & (merged_df['sale_date'] <= merged_df['promo_end_date'])

```

```python
# Update effective_price if the sale was within the promo period
merged_df['effective_price'] = merged_df.apply(
    lambda row: row['sale_price_time_promo'] if row['on_promo'] else row['effective_price'], axis=1
)

```

```python
# Recalculate sum_total with updated prices
merged_df['sum_total'] = merged_df['effective_price'] * merged_df['quantity']

```

```python
# Drop unnecessary columns
discount_sale_df = merged_df.drop(columns=['promo_end_date', 'on_promo'])

# Check result
discount_sale_df.head()
```

```python
discount_sale_df[['effective_price', 'quantity', 'sum_total', 'sale_date', 'date', 'number_disc_day', 'sale_price_before_promo', 'sale_price_time_promo', 'channel']].sample(10)
```

```python
discount_sale_df.columns
```

```python
discount_sale_df[['sale_price_before_promo', 'sale_price_time_promo']] = discount_sale_df[['sale_price_before_promo', 'sale_price_time_promo']].fillna(0)
```

```python
import numpy as np

discount_sale_df['discount'] = np.where(
    discount_sale_df['sale_price_time_promo'] > discount_sale_df['sale_price_before_promo'],
    0,  # Set discount to 0 if the condition is met
    discount_sale_df['sale_price_before_promo'] - discount_sale_df['sale_price_time_promo']
)
```

```python
discount_sale_df[discount_sale_df['discount'] < 0]
```

```python
final_discount_sale = discount_sale_df[['item_id', 'store_id', 'effective_price', 'discount', 'quantity', 'sum_total', 'sale_date', 'channel']]
final_discount_sale.sample(10)
```

####  Decompose the Time Series (Trend, Seasonality, Residuals)

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Ensure sale_date is datetime
data['sale_date'] = pd.to_datetime(data['sale_date'])

# Set date as index
data = data.set_index('sale_date')

# Resample to daily frequency and fill missing values with 0
daily_sales = data['quantity'].resample('D').sum().fillna(0)

# Perform seasonal decomposition
decomposition = seasonal_decompose(daily_sales, model='additive')

# Plot results
fig, axes = plt.subplots(4, 1, figsize=(10, 8), sharex=True)
decomposition.observed.plot(ax=axes[0], title='Observed')
decomposition.trend.plot(ax=axes[1], title='Trend')
decomposition.seasonal.plot(ax=axes[2], title='Seasonality')
decomposition.resid.plot(ax=axes[3], title='Residuals')

plt.tight_layout()
plt.show()
```

#### Rolling Average Smoothing (Highlight Long-Term Trends)
